```python
from collections import defaultdict

class Graph:
    def __init__(self, vertices):
        self.graph = defaultdict(list)
        self.V = vertices

    def add_edge(self, u, v):
        self.graph[u].append(v)
        self.graph[v].append(u)

    def find_euler(self):
        # Hitung derajat ganjil
        odd = [v for v in range(self.V) if len(self.graph[v]) % 2 != 0]

        # Tentukan simpul awal
        if len(odd) == 0:
            start = 0
            print("Sirkuit Euler ditemukan:")
        elif len(odd) == 2:
            start = odd[0]
            print("Lintasan Euler ditemukan:")
        else:
            print("Tidak ada lintasan/sirkuit Euler.")
            return

        stack = [start]
        path = []

        while stack:
            v = stack[-1]
            if self.graph[v]:
                u = self.graph[v][-1]
                # Hapus sisi dua arah
                self.graph[v].remove(u)
                self.graph[u].remove(v)
                stack.append(u)
            else:
                path.append(stack.pop())

        print(" -> ".join(map(str, path[::-1])))


# ======== Contoh Penggunaan ========

g = Graph(5)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(3, 4)
g.add_edge(4, 2)

print("Graf 1:")
g.find_euler()
```
